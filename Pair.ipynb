```python
df_list_formal = pd.DataFrame()
df_list_informal = pd.DataFrame()

z = list(df.columns)
for i,j in enumerate(z) :
  if i % 2 == 0 :
    # print(j)
    df_list_formal = pd.concat([df_list_formal, df[str(j)]])
  else:
    df_list_informal = pd.concat([df_list_informal, df[str(j)]])
```

```python
final_pair = pd.concat([df_list_formal, df_list_informal], axis=1, ignore_index=True)
```

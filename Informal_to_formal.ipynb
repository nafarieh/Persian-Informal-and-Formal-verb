```python
# coding: utf-8

from __future__ import unicode_literals
from nltk.stem.api import StemmerI


class Stemmer(StemmerI):
	"""
	>>> stemmer = Stemmer()
	>>> stemmer.stem('کتابی')
	'کتاب'
	>>> stemmer.stem('کتاب‌ها')
	'کتاب'
	>>> stemmer.stem('کتاب‌هایی')
	'کتاب'
	>>> stemmer.stem('کتابهایشان')
	'کتاب'
	>>> stemmer.stem('اندیشه‌اش')
	'اندیشه'
	>>> stemmer.stem('خانۀ')
	'خانه'
	"""
	# استثنا
  #خونمون - سفرمون- کاشانمون- 
	#سراتون , فالووراتون , دراتون  , 


	def __init__(self):
		self.ends = ['‌','مون','تون','شون'
			          # ,  'ش'
								, 'هام', 'هات', 'هاش', 'هامون', 'هاتون' , 'هاشون'
			          ,'ام'
								,'ات'
								,'اش'
								, 'مون'
								, 'تون'
								,'شون'
								,'مونو'
								,'تونو'
								,'شونو'
								# , 'ا'
								]

		self.except_list = [
										'دریا' ,'انا','یالا','بالا','راستا','آیا'
										,'تا', 'با' , 'یا' , 'ما' , 'شما',
										'ندا' ,'زهرا', 'دنیا','رویا' , 'رها'
										,'اینجا','آنجا','هرجا','همه جا','حتما','دقیقا','نسبتا','قاعدتا','اونجا'
										# ,'درخشش','درفش'
									]
		
	def stem(self, word):
		word_2 = ""
		for end in self.ends :
			if word not in self.except_list :
				# print(end)
				if word.endswith(end):
					word = word[:-len(end)]
					word_2 = end

		if word.endswith('ۀ'):
			word = word[:-1] + 'ه'

		return word,word_2
```

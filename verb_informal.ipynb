```python
# bon_mazi = ["آب داد" , "آرایش کرد" ,"آسیاب کرد","اومد"]
bon_mazi = df_verb["Bon_mazi"] 
zamir = ["م" , "ی" , "" , "یم" , "ین" , "ن"]

df = pd.DataFrame()


zaman_mazi_sadeh = []
zaman_mazi_estemrari = []
zaman_mazi_baeid = []
zaman_mazi_eltezami = []
mazi_mostamar = []

# zaman_mazi = bon_mazi 
# zaman_mazi_naghli = zaman_mazi
# zaman_naghli_estem =zaman_mazi_estemrari 
# zaman_ayandeh = zaman_mazi_estemrari
# zaman_mazi_abad = zaman_mazi_baeid
##########################
print("zaman_mazi_sadeh")
for my_bon_mazi in bon_mazi :  
  for my_zamir in zamir :
    if len(my_bon_mazi.split()) >1 :
          zaman_mazi_sadeh.append(str(' '.join(my_bon_mazi.split()[:-1]))  + " " + str(my_bon_mazi.split()[-1]) +  my_zamir)
          print(str(' '.join(my_bon_mazi.split()[:-1]))  + " " + str(my_bon_mazi.split()[-1]) + my_zamir)
    else:
      zaman_mazi_sadeh.append(my_bon_mazi+ my_zamir)
      print( my_bon_mazi+ my_zamir)
df['zaman_mazi_sadeh'] = zaman_mazi_sadeh
##########################
print("zaman_mazi_estemrari")
for my_bon_mazi in bon_mazi :  
  for my_zamir in zamir :
    if len(my_bon_mazi.split()) >1 :
          zaman_mazi_estemrari.append(str(' '.join(my_bon_mazi.split()[:-1]))  +  " می " + str(my_bon_mazi.split()[-1])+ my_zamir)
          print(str(' '.join(my_bon_mazi.split()[:-1]))  +  " می " + str(my_bon_mazi.split()[-1])+ my_zamir)
    else:
      zaman_mazi_estemrari.append("می " + my_bon_mazi+ my_zamir)
      print(" می " + my_bon_mazi+ my_zamir)
df['zaman_mazi_estemrari'] = zaman_mazi_estemrari
##########################
print("zaman_mazi_baeid")
for my_bon_mazi in bon_mazi : 
  for my_zamir in zamir :
    # zaman_mazi_baeid +=  ["ه بود"] + my_bon_mazi
    if len(my_bon_mazi.split()) >1 :
          zaman_mazi_baeid.append(str(' '.join(my_bon_mazi.split()[:-1]))  +" " + str(my_bon_mazi.split()[-1]) + "ه بود"+ my_zamir)
          print(str(' '.join(my_bon_mazi.split()[:-1]))  +" " + str(my_bon_mazi.split()[-1]) + "ه بود"+ my_zamir)
    else: 
      zaman_mazi_baeid.append(my_bon_mazi + "ه بود"+ my_zamir)
      print( my_bon_mazi + "ه بود"+ my_zamir)
df['zaman_mazi_baeid'] = zaman_mazi_baeid
###########################
print("zaman_mazi_eltezami")

for my_bon_mazi in bon_mazi : 
  for my_zamir in zamir :    
    # zaman_mazi_eltezami += ["ه باش"] + my_bon_mazi
    if len(my_bon_mazi.split()) >1 :
          zaman_mazi_eltezami.append(str(' '.join(my_bon_mazi.split()[:-1]))  + " " + str(my_bon_mazi.split()[-1]) + "ه باش" + my_zamir)
          print(str(' '.join(my_bon_mazi.split()[:-1]))  + " " + str(my_bon_mazi.split()[-1]) + "ه باش" + my_zamir)
    else: 
      zaman_mazi_eltezami.append(my_bon_mazi + "ه باش" + my_zamir)
      print(my_bon_mazi + "ه باش" + my_zamir)
df['zaman_mazi_eltezami'] = zaman_mazi_eltezami
#####################################################################
# bon_mozare = ["آب د" , "آرایش کن" ,"آسیاب کن","یا"]
bon_mozare =  df_verb["Bon_mozare"] 
zamir = ["م" , "ی" , "ه" , "یم" , "ین" , "ن"]

mozare_mostamar = []
mozare_ekhbari = []
mozare_eltezami = []
mozare_estemrari = []
Ayandeh = []

print("mozare mostamar")
pish_mozare_mostamar = []
for my_bon_mozare in bon_mazi : 
  for my_zamir in zamir :
    # pish_mazi_mostamar = ["داشت"]
    my_pish_mozare_mostamar= " داشت" + my_zamir + " "
    
    if len(my_bon_mozare.split()) >1 :
          mozare_mostamar.append(my_pish_mozare_mostamar + str(' '.join(my_bon_mozare.split()[:-1])) +  " می " + str(my_bon_mozare.split()[-1])+ my_zamir)
          print(my_pish_mozare_mostamar + str(' '.join(my_bon_mozare.split()[:-1])) +  " می " + str(my_bon_mozare.split()[-1])+ my_zamir)
    else:
      mozare_mostamar.append(my_pish_mozare_mostamar + " می " + my_bon_mozare+ my_zamir)
      print(my_pish_mozare_mostamar + " می " + my_bon_mozare+ my_zamir)
df['mazi_mostamar'] = mozare_mostamar
# mazi_mostamar_naghli = mazi_mostamar
#___________________________________________________

###############################
print("mozare_ekhbari")
for my_bon_mozare in bon_mozare : 
  for my_zamir in zamir :  
    if len(my_bon_mozare.split()) >1 :
          mozare_ekhbari.append(str(' '.join(my_bon_mozare.split()[:-1])) +  " می " + str(my_bon_mozare.split()[-1]) + my_zamir)
          print(str(' '.join(my_bon_mozare.split()[:-1])) +  " می " + str(my_bon_mozare.split()[-1]) + my_zamir)
    else:
      mozare_ekhbari.append(" می " + my_bon_mozare + my_zamir)
      print(" می " + my_bon_mozare + my_zamir)
df['mozare_ekhbari'] = mozare_ekhbari
##############################

print("mozare_eltezami")
for my_bon_mozare in bon_mozare : 
  for my_zamir in zamir :  
    if len(my_bon_mozare.split()) >1 :
          mozare_eltezami.append(str(' '.join(my_bon_mozare.split()[:-1])) +  " ب" + str(my_bon_mozare.split()[-1])  + my_zamir)
          print(str(' '.join(my_bon_mozare.split()[:-1])) +  " ب" + str(my_bon_mozare.split()[-1])  + my_zamir)
    else:
      mozare_eltezami.append(" ب" + my_bon_mozare + my_zamir)
      print(" ب" + my_bon_mozare + my_zamir)
df['mozare_eltezami'] = mozare_eltezami

##############################
# mozare_estemrari =  pish_mozare_estemrari + mozare_ekhbari

print("mozare_estemrari")
pish_mazi_mostamar = []
for my_bon_mozare in bon_mozare :  
  for my_zamir in zamir :
    pish_mozare_estemrari= "دار" + my_zamir +" "
    
    if len(my_bon_mozare.split()) >1 :
          mozare_estemrari.append(pish_mozare_estemrari + str(' '.join(my_bon_mozare.split()[:-1])) +  " می " + str(my_bon_mozare.split()[-1]) + my_zamir)
          print(pish_mozare_estemrari + my_bon_mozare.split()[0] +  " می " + str(my_bon_mozare.split()[-1]) + my_zamir)
    else:
      mozare_estemrari.append(pish_mozare_estemrari + " می " + my_bon_mozare+ my_zamir)
      print(pish_mozare_estemrari + " می " + my_bon_mozare+ my_zamir)
df['mozare_estemrari'] = mozare_estemrari
##############################
print("Ayandeh")
for my_bon_mozare in bon_mozare : 
  for my_zamir in zamir :  
    if len(my_bon_mozare.split()) >1 :   
          Ayandeh.append(" می خوا" + my_zamir + " " + str(' '.join(my_bon_mozare.split()[:-1])) +" "+ "ب" +str(my_bon_mozare.split()[-1])  + my_zamir)
          print(" می خوا" + my_zamir + " " + str(my_bon_mozare.split()[-1])  +" "+ "ب" +  str(my_bon_mozare.split()[-1])  + my_zamir)
    else:
      Ayandeh.append(" می خوا"  + my_zamir  +" " + "ب" + my_bon_mozare + my_zamir)
      print("می خوا"  + my_zamir  +" "+ "ب" + my_bon_mozare + my_zamir)
df['َAyandeh'] = Ayandeh
##############################
print("Amr")
bon_mozare =  df_verb["Bon_mozare"] 
# zamir = ["ه" , "ه" , "یم" , "ین" , "ن" , "-"]
zamir = ["-" , "" , "ه" , "یم" , "ین" , "ن"]
Amr = []

for my_bon_mozare in bon_mozare : 
  for my_zamir in zamir :  
    if my_zamir == '-':
      Amr.append("-")
    else:
      if len(my_bon_mozare.split()) >1 :
            Amr.append(str(' '.join(my_bon_mozare.split()[:-1])) +  " ب" + str(my_bon_mozare.split()[-1])  + my_zamir)
            print(str(' '.join(my_bon_mozare.split()[:-1])) +  " ب" + str(my_bon_mozare.split()[-1])  + my_zamir)
      else:
        Amr.append(" ب" + my_bon_mozare + my_zamir)
        print(" ب" + my_bon_mozare + my_zamir)
df['Amr'] = Amr

##############################

df.to_csv('3_Informal_dataframe_verb.csv', encoding='utf-8')
```
